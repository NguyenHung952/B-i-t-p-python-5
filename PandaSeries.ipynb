```python
#Câu 1:

import numpy as np
import pandas as pd
arr_1 = np.array([2, 4, 6, 8, 10])
arr_2 = np.array([1, 3, 5, 7, 11])
ser1 = pd.Series(arr_1)
ser2 = pd.Series(arr_2)

#a/Tạo biến kiểu Serries ser1 từ arr_1, ser2 từ arr_2. In danh sách các phần tử của ser1 và ser2
print("danh sách các phần tử của ser1:")
print(ser1)
print("danh sách các phần tử của ser2:")
print(ser2)
#b/ Thực hiện phép toán và thể hiện kết quả của:  ser1 + ser2 
print("Phep cong cua series:")
sera = ser1 + ser2
print(sera)
#c/ Thực hiện phép toán và thể hiện kết quả của:  ser1 - ser2 
print("Phep tru cua series:")
serb = ser1 - ser2
print(serb)
#d/ Thực hiện phép toán và thể hiện kết quả của:  ser1 * ser2
print("Phep nhan cua series:")
serc = ser1 * ser2
print(serc)
#e/ Thực hiện phép toán và thể hiện kết quả của:  ser1 / ser2
print("Phep chia cua series:")
serd = ser1 - ser2
print(serd)

#Câu 3:
#a/ Thêm 2 phần tử [6, 12] vào ser2. In lại danh sách các phần tử của ser2.
print("Phan tu cua ser 2 sau khi cap nhat:")
ser2 = pd.Series(np.append(ser2,[6, 12]))
print(ser2)
#b/ Tạo series ser3 chỉ chứa các phần tử có trong ser1 mà không có trong ser2. 
#In danh sách các phần tử của ser3
arr_3 = []
[ arr_3.append(x) for x in ser1 if x not in ser2 ]
ser3 = pd.Series(arr_3)
print("danh sách các phần tử của ser3: ")
print(ser3)
#c/ Tạo series ser4 chỉ chứa các phần tử có trong ser2 mà không có trong ser1. 
#In danh sách các phần tử của ser4
arr_4 = []
[ arr_4.append(x) for x in ser2 if x not in ser1 ]
ser4 = pd.Series(arr_4)
print("danh sách các phần tử của ser4: ")
print(ser4)
```

```python
# Câu 1:
import numpy as np
import pandas as pd
#a/ Tạo series ser6 có 35 phần tử số nguyên ngẫu nhiên có giá trị trong khoảng từ 1 đến 9. Cho biết kích thước
#(shape) của ser6. Xem 5 dòng dữ liệu đầu tiên (head) và 5 dòng dữ liệu cuối cùng (tail) có trong ser6

arr_6 = np.random.randint(1,10,35)
ser6 = pd.Series(arr_6)
print(ser6)
print("5 dòng dữ liệu đầu tiên có trong ser6:")
print(ser6.head(5))
print("5 dòng dữ liệu cuối cùng có trong ser6:")
print(ser6.tail(5))

#b/ In danh sách các phần tử của ser6 theo dạng array

print("danh sách các phần tử của ser6 theo dạng array:")
print(arr_6)

#c/ Cho biết thông tin thống kê chung (describe()) của ser6

print("thông tin thống kê chung của ser6:")
print(ser6.describe())

#d/ Cho biết tổng của các phần tử có trong ser6

print("tổng của các phần tử có trong ser6:")
print(ser6.sum())

#e/ Cho biết phần tử có tần suất xuất hiện nhiều nhất trong ser6

print("phần tử có tần suất xuất hiện nhiều nhất trong ser6:")
print(ser6.mode())

#Câu 2:  Liệt kê các dòng trong ser6 mà giá trị chia hết cho 2 và cho 3

print("các dòng trong ser6 mà giá trị chia hết cho 2 và cho 3:")

answer = []

for element in ser6:
    if element % 2 == 0 and element % 3 ==0:
        answer.append(element)
if len(answer) == 0:
    answer = [0]
    
print(answer)

#Câu 3: In các phần tử ở vị trí 0, 5, 10, 15 trong ser6

print("In các phần tử ở vị trí 0, 5, 10, 15 trong ser6:")
print(ser6[0])
print(ser6[5])
print(ser6[10])
print(ser6[15])

#Câu 4: In ra các giá trị unique (array) trong ser6

print("In ra các giá trị unique (array) trong ser6:")
unique = np.unique(arr_6)
print(unique)

#Câu 5: Tạo series ser7 với mỗi phần tử có giá trị = lập phương của phần tử trong ser6. 
#Xem 5 dòng dữ liệu đầu tiên (head) của ser7
ser7 = np.array(ser6)
print("ser7 =")
print(ser7)
print("5 dòng dữ liệu đầu tiên (head) của ser7:")
print(ser7[:5])
```
